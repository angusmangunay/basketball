```python
import pandas as pd
pd.set_option('display.max_columns', None)

df = pd.read_csv("cbb.csv")
df.head()
```

```python
#New column if made postseason

def ps(row):
    if row['POSTSEASON'] != "NA":
        return 1
    else:
        return 0 
    
df['PLAYOFF'] = df.apply(lambda row: ps(row), axis=1)
```
